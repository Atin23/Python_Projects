```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text'])\
        .size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

## Observations:
1. When Rating is between 4.5 to 4.9---> Excellent
2. When Rating are between 4.0 to 3.4---> Very good
3. when Rating is between 3.5 to 3.9----> Good
4. when Rating is between 2.5 to 3.4----> Average
5. when Rating is between 1.8 to 2.4----> Poor

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12, 6)
sns.barplot(data=ratings, x="Aggregate rating", y="Rating Count", width=1, hue='Rating color',\
            palette=['blue','red','orange','yellow','green','green'])
plt.show()
```

### Observation:
1. Not Rated count is very high
2. Maximum number of ratings are between 2.5 to 3.4

```python
## Count plot
sns.countplot(x="Rating color",data=ratings,palette=['blue','red','orange','yellow','green','green'])
plt.show()
```
